```python
from stdash.data import load_data
import pandas as pd

data = load_data()
df = pd.DataFrame(data)


df['request_time'] = pd.to_datetime(df['request_time']).dt.floor('30min')
df['prediction_time'] = pd.to_datetime(df['prediction_time']).dt.floor('30min')

df.head(3)
```

```python
a_counts = df.groupby('request_time').size().reset_index(name='count')
a_counts['type'] = 'request_time'
a_counts.rename(columns={'request_time': 'yymmdd'}, inplace=True)

# B 열에 대해서 count
b_counts = df.groupby('prediction_time').size().reset_index(name='count')
b_counts['type'] = 'prediction_time'
b_counts.rename(columns={'prediction_time': 'yymmdd'}, inplace=True)

# A와 B count 데이터 합치기
result = pd.concat([a_counts, b_counts]).sort_values(by='yymmdd').reset_index(drop=True)

# yymmdd를 'yymmdd' 형식으로 변환
result['yymmdd'] = result['yymmdd'].dt.strftime('%y%m%d')

print(result)
```

```python
result_df = df.groupby(['request_time', 'request_user']).size().reset_index(name='count')
result_df
```

```python
grouped_A  = df.groupby('request_time').size().reset_index(name='count')
grouped_A['type'] = 'request_time'
grouped_B = df.groupby('prediction_time').size().reset_index(name='count')
grouped_B['type'] = 'prediction_time'

# 컬럼명을 동일하게 맞추기 위해 request_time, prediction_time을 'time'으로 변경
result_df = pd.concat([grouped_A, grouped_B], ignore_index=True)
result_df = result_df.fillna(0)

result_df.rename(columns={'index': 'date'}, inplace=True)

result_df
```

```python
import pandas as pd

# 예시 데이터 생성
data = {'A': pd.to_datetime(['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-03', '2023-01-03']),
        'B': pd.to_datetime(['2023-01-02', '2023-01-02', '2023-01-03', '2023-01-03', '2023-01-04'])}
df = pd.DataFrame(data)

# A 열에 대해서 count
a_counts = df.groupby('A').size().reset_index(name='count')
a_counts['type'] = 'A'
a_counts.rename(columns={'A': 'yymmdd'}, inplace=True)

# B 열에 대해서 count
b_counts = df.groupby('B').size().reset_index(name='count')
b_counts['type'] = 'B'
b_counts.rename(columns={'B': 'yymmdd'}, inplace=True)

# A와 B count 데이터 합치기
result = pd.concat([a_counts, b_counts]).sort_values(by='yymmdd').reset_index(drop=True)

# yymmdd를 'yymmdd' 형식으로 변환
result['yymmdd'] = result['yymmdd'].dt.strftime('%y%m%d')

print(result)
```
